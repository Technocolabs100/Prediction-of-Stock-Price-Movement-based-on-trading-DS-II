```python
# Importing Necessary Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 
import datetime
```

```python
# Reading Data File
df1=pd.read_csv(r"C:\Users\Priti\Downloads\Technocolabs Internship Program\DataFrame.csv")
```

```python
#For checking shape of dataset
print(df1.shape)
```

```python
# To print top 5 lines of the dataset
df1.head()
```

```python
# Here data contains one unnecessary column Unnamed:7 so removing this
df1.drop('Unnamed: 7', axis=1, inplace=True)
```

```python
df1.head()
```

```python
# Parsing the date in proper format as year-month-day
df1['Date']=pd.to_datetime(df1["Date"], format="%Y%m%d")
df1.head()
```

```python
# Checking the dataset contain missing values or not 
df1.isnull().values.any()
```

```python
# finding sum of missing values
df1.isnull().sum()
```

```python
# To finding the basic information about the dataset
df1.info()
```

```python
# To checking the dataset contain any duplicated values or not
df1[df1.duplicated()]
```

```python
# To find Descriptive of the dataset
df1.describe()
```

```python
# Adding some extra columns for the better understanding the dataset
df1['profit/loss']=df1['open']-df1['close']
df1['max_profit']=df1['high']-df1['low']
df1.head()
```

```python
df1['profit/loss'].describe()
```

```python
# To ploting the distribution of profit or loss 
sns.distplot(df1['profit/loss'],kde=False,bins=50) # safe to ignore the warnings
```

```python
# To plot the pairplot of Dataset
sns.pairplot(df1)
```

```python
# Plotting Heatmap of dataset
sns.heatmap(df1.corr())
```
