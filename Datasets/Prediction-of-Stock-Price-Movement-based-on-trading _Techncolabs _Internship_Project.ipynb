# EDA ANALYSIS ON DATAFRAME.CSV

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data = data.set_index(['TYPE','DATE','TIME'])
```

```python
features = ['OPEN', 'HIGH', 'LOW', 'CLOSE']
list(enumerate(features))
```

```python
plt.figure(figsize=(13,10))
for i in enumerate(features):
    plt.subplot(2,2,i[0]+1)
    base_color = sns.color_palette()[i[0]]
    sns.distplot(x = data[i[1]] , kde = False , color = base_color , hist_kws={'alpha':1});
    plt.xlabel(i[1])
    plt.title(f'Distribution Plot for {i[1]}')
```

```python
g = sns.PairGrid(data)
g.map_upper(sns.scatterplot,color='red')
g.map_lower(sns.scatterplot, color='orange')
g.map_diag(plt.hist)
```

```python
feature1 = ['OPEN', 'HIGH', 'LOW', 'CLOSE']
feature2 = ['OPEN', 'HIGH', 'LOW', 'CLOSE']
for i in enumerate(feature1):
    for j in enumerate(feature2):
        if (i[1] != j[1]):
            plt.figure(figsize=(5,3))
            base_color = sns.color_palette()[i[0]]
            sns.scatterplot(x = data[i[1]], y = data[j[1]] , color = base_color);
```

```python
data = data.reset_index()
data = data.set_index('DATE')
plt.figure(figsize = (12 , 6))
sns.lineplot(  data = data )
plt.xticks(rotation = 90);
plt.title("LINE PLOT W.R.T. DATE")
plt.xlabel("DATE")
```

```python
sns.heatmap(data.corr(), vmin=-1, vmax=1, annot=True , cmap = 'Paired_r')
```

```python
plt.figure(figsize=(13,10))
for i in enumerate(features):
    plt.subplot(2,2,i[0]+1)
    base_color = sns.color_palette()[i[0]]
    sns.violinplot(x = data[i[1]] , kde = False , color = base_color);
    plt.title(f'Violin Plot for {i[1]}')
```
