# EDA ANALYSIS ON DATAFRAME.CSV

```python
import matplotlib.pyplot as plt        #IMPORTING LIBRARIES FOR VISUALIZATION
import seaborn as sns 
```

```python
data = data.set_index(['TYPE','DATE','TIME'])
```

```python
features = ['OPEN', 'HIGH', 'LOW', 'CLOSE' , 'Profit/Loss' , 'Max_Profit']
list(enumerate(features))
```

```python
plt.figure(figsize=(15,15))
for i in enumerate(features):
    plt.subplot(3,3,i[0]+1)                         #DISTRIBUTION PLOTS FOR EACH ATTRIBUTE
    base_color = sns.color_palette()[i[0]]
    sns.distplot(x = data[i[1]] , kde = False , color = base_color , hist_kws={'alpha':1});
    plt.xlabel(i[1])
    plt.title(f'Distribution Plot for {i[1]}')
```

```python
g = sns.PairGrid(data)
g.map_upper(sns.scatterplot,color='red')              #SCATTER PLOT B/W DIFFERENT ATTRIBUTES
g.map_lower(sns.scatterplot, color='orange')
g.map_diag(plt.hist)
```

```python
feature1 = ['OPEN', 'HIGH', 'LOW', 'CLOSE' ,'Profit/Loss' , 'Max_Profit']
feature2 = ['OPEN', 'HIGH', 'LOW', 'CLOSE' , 'Profit/Loss' , 'Max_Profit']
for i in enumerate(feature1):                        #SCATTER PLOT USING NESTED LOOP
    for j in enumerate(feature2):
        if (i[1] != j[1]):
            plt.figure(figsize=(5,3))
            base_color = sns.color_palette()[i[0]]
            sns.scatterplot(x = data[i[1]], y = data[j[1]] , color = base_color);
```

```python
data = data.reset_index()
data = data.set_index('DATE')
plt.figure(figsize = (12 , 6))
sns.lineplot(  data = data )                     # LINE PLOT FOR ALL ATTRIBUTES ACC. TO DATE
plt.xticks(rotation = 90);
plt.title("LINE PLOT W.R.T. DATE")
plt.xlabel("DATE")
```

```python
sns.heatmap(data.corr(), vmin=-1, vmax=1, annot=True , cmap = 'Paired_r') #HEATMAP FOR CORRELATION B/W ATTRIBUTES
```

```python
plt.figure(figsize=(15,15))
for i in enumerate(features):
    plt.subplot(3,3,i[0]+1)                               #VIOLIN PLOT FOR ALL ATTRIBUTES
    base_color = sns.color_palette()[i[0]]
    sns.violinplot(x = data[i[1]] , kde = False , color = base_color);
    plt.title(f'Violin Plot for {i[1]}')
```

# EDA ANALYSIS ON MSFT.CSV

```python
data_1 = data_1.set_index(['DATE'])
```

```python
features = ['OPEN', 'HIGH', 'LOW', 'CLOSE' , 'ADJ CLOSE' , 'VOLUME' , 'Profit/Loss' , 'Max_Profit']
list(enumerate(features))
```

```python
plt.figure(figsize=(18,20))
for i in enumerate(features):                         #DISTRIBUTION PLOT FOR ALL ATTRIBUTES
    plt.subplot(4,4,i[0]+1)
    base_color = sns.color_palette()[i[0]]            
    sns.distplot(x = data_1[i[1]] , kde = False , color = base_color , hist_kws={'alpha':1});
    plt.xlabel(i[1])
    plt.title(f'Distribution Plot for {i[1]}')
```

```python
sns.pairplot(data = data_1)   #SCATTER PLOT BETWEEN EACH ATTRIBUTES
```

```python
plt.figure(figsize = (12 , 6))
sns.lineplot(  data = data_1   )
plt.xticks(rotation = 90);                   #LINE PLOT FOR ALL ATTRIBUTE ACC. TO DATE
plt.title("LINE PLOT W.R.T. DATE")
plt.xlabel("DATE")
```

```python
sns.heatmap(data_1.corr(), vmin=-1, vmax=1, annot=True , cmap = 'YlGnBu')    #HEATMAP FOR CORRELATION B/W ATTRIBUTES
```

```python
plt.figure(figsize=(15,20))
for i in enumerate(features):
    plt.subplot(4,4,i[0]+1)                             #BOX PLOT FOR EACH ATTRIBUTES
    base_color = sns.color_palette()[i[0]]
    sns.boxplot(x = data_1[i[1]] , color = base_color);
    plt.title(f'Box Plot for {i[1]}')
```
