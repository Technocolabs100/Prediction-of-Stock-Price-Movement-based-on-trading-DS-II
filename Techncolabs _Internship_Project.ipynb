# MODELLING ON DATAFRAME.CSV USING LSTM ALGORITHM

```python
data.drop(['Profit/Loss' , 'Max_Profit'] , axis = 1 , inplace = True) #dropping the profit/loss and max_profit column for training and testing
```

```python
df = data.reset_index()['CLOSE']             #creating seperate variable for target attribute'CLOSE'
df.shape 
```

```python
import matplotlib.pyplot as plt                             # Trend for 'CLOSE'
plt.plot(df)
plt.title('TREND FOR CLOSE IN ACTUAL FORMAT')
```

```python
from sklearn.preprocessing import MinMaxScaler                        # Scaling 'CLOSE' B/W 0 and 1
scaler = MinMaxScaler(feature_range=(0,1))
df = scaler.fit_transform(np.array(df).reshape(-1,1))
```

```python
df  # scaled 'CLOSE' 
```

```python
train_size = int(len(df)*0.65)                             #splitting dataset into train and test 
test_size = len(df)-train_size
train_data,test_data = df[0:train_size,:],df[train_size:len(df),:1]
```

```python
train_size,test_size        #shape of spilted train and test data
```

```python
import numpy
def create_dataset(dataset, time_step=1):               # convert an array of values into a dataset matrix
    dataX, dataY = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step), 0]   ###i=0, 0,1,2,3 and so on .............. 
        dataX.append(a)
        dataY.append(dataset[i + time_step, 0])
    return numpy.array(dataX), numpy.array(dataY)
```

```python
time_step = 100
X_train, Y_train = create_dataset(train_data, time_step)      # reshape into X=t,t+1,t+2,t+3 and Y=t+4
X_test, Y_test = create_dataset(test_data, time_step)
```

```python
print(X_train)         # similar way we can print Y_train and X_test ,Y_test
```

```python
print(X_train.shape), print(Y_train.shape)     #shape of trained data
```

```python
print(X_test.shape), print(Y_test.shape)         #shape of test data
```

```python
#reshaping input to be [samples, time steps, features] which is required for LSTM
X_train =X_train.reshape(X_train.shape[0],X_train.shape[1] , 1)
X_test = X_test.reshape(X_test.shape[0],X_test.shape[1] , 1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense        #importing libraries for LSTM MODEL
from tensorflow.keras.layers import LSTM
```

```python
model=Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(100,1)))
model.add(LSTM(50,return_sequences=True))                               # MODEL PREPARED FOR LSTM
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model.summary()          #Summary of Model Prepared
```

```python
model.fit(X_train,Y_train,validation_data=(X_test,Y_test),epochs=100,batch_size=64,verbose=1)    # fitting Model
```

```python
train_predict=model.predict(X_train)                            # predicting Model
test_predict=model.predict(X_test)                                    
```

```python
import math
from sklearn.metrics import mean_squared_error
train = math.sqrt(mean_squared_error(Y_train,train_predict))
test =  math.sqrt(mean_squared_error(Y_test,test_predict))                # Calculating Root Mean Square Error
print("Root Mean Square Error for Trained Data is :" , train)
print("Root Mean Square Error for Test Data is :" , test)
```

```python
from sklearn.metrics import r2_score
Acc_tr = r2_score( Y_train , train_predict)                                    # Calculating R2 SCORE 
Acc_te = r2_score( Y_test , test_predict)
print("R2 Score for Trained Model is :" , Acc_tr)
print("R2 Score for Test Model is :" , Acc_te)
```

```python
Y_train = np.array(Y_train).reshape(-1,1)
Y_train = scaler.inverse_transform(Y_train)
Y_train = Y_train.flatten()                                             # Transforming into Original Form Y_Train
Y_train = Y_train.astype(float)
Y_train
```

```python
arr1 = np.array(train_predict).reshape(-1,1)
arr1 = scaler.inverse_transform(arr1)
arr1 = arr1.flatten()                                           # Transforming into Original Form train_predict
arr1 = arr1.astype(float)
arr1
```

```python
Y_test = np.array(Y_test).reshape(-1,1)
Y_test = scaler.inverse_transform(Y_test)
Y_test = Y_test.flatten()                                             # Transforming into Original Form Y_Test
Y_test = Y_test.astype(float)
Y_test
```

```python
arr2 = np.array(test_predict).reshape(-1,1)
arr2 = scaler.inverse_transform(arr2)                          # Transforming into Original Form test_predict
arr2 = arr2.flatten()
arr2 = arr2.astype(float)
arr2
```

```python
pd.DataFrame({'ACTUAL_VALUE_Train':Y_train , 'PREDICTED_VALUE_Train':arr1})  # Dataframe Created for Trained (ACTUAL & PREDICTED) Value
```

```python
pd.DataFrame({'ACTUAL_VALUE_Test':Y_test , 'PREDICTED_VALUE_Test':arr2})    # Dataframe Created for Test (ACTUAL & PREDICTED) Value
```

```python
plt.plot(Y_train, label='ACTUAL' , color = 'blue')
plt.plot(arr1, label='LSTM PREDICTED' , color = 'red')
plt.title("LSTM's_Prediction on Training")                          # Graph For Trained
plt.legend()
plt.show()
```

```python
plt.plot(Y_test, label='ACTUAL' , color = 'blue')
plt.plot(arr2, label='LSTM PREDICTED' , color = 'red')
plt.title("LSTM's_Prediction on Testing")                            # GRAPH FOR TEST
plt.legend()
plt.show()
```

```python
#reshaping input to be [samples, time steps, features] which is required for GRU
X_train =X_train.reshape(X_train.shape[0],X_train.shape[1] , 1)
X_test = X_test.reshape(X_test.shape[0],X_test.shape[1] , 1)
```

```python
from keras.layers import  Dropout, GRU
```

```python
# The GRU architecture
regressorGRU = Sequential()
# First GRU layer with Dropout regularisation
regressorGRU.add(GRU(units=50, return_sequences=True, input_shape=(X_train.shape[1],1)))
regressorGRU.add(Dropout(0.2))
# Second GRU layer
regressorGRU.add(GRU(units=50, return_sequences=True))
regressorGRU.add(Dropout(0.2))
# Third GRU layer
regressorGRU.add(GRU(units=50, return_sequences=True))
regressorGRU.add(Dropout(0.2))
# Fourth GRU layer
regressorGRU.add(GRU(units=50))
regressorGRU.add(Dropout(0.2))
# The output layer
regressorGRU.add(Dense(units=1))

# Compiling the RNN
regressorGRU.compile(optimizer='adam',loss='mean_squared_error')
# Fitting to the training set
regressorGRU.fit(X_train,Y_train,epochs=50,batch_size=1000)
```

```python
train_predict=regressorGRU.predict(X_train)                            # predicting Model
test_predict=regressorGRU.predict(X_test)
```

```python
import math
from sklearn.metrics import mean_squared_error
train = math.sqrt(mean_squared_error(Y_train,train_predict))
test =  math.sqrt(mean_squared_error(Y_test,test_predict))                      # Calculating Root Mean Square Error
print("Root Mean Square Error for Trained Data is :" , train)
print("Root Mean Square Error for Test Data is :" , test)
```

```python
from sklearn.metrics import r2_score
Acc_tr = r2_score( Y_train , train_predict)                                    # Calculating R2 SCORE 
Acc_te = r2_score( Y_test , test_predict)
print("R2 Score for Trained Model is :" , Acc_tr)
print("R2 Score for Test Model is :" , Acc_te)
```
